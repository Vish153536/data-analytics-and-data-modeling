```python
from keras.models import Sequential
from keras.layers import Dense , Dropout
from keras.callbacks import EarlyStopping
from keras.utils.np_utils import to_categorical

X=datas.iloc[:,:8]
Y=datas.iloc[:,8]
```

```python
model=Sequential()
model.add(Dense(12,input_shape=(X.shape[1],),activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(3,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss="categorical_crossentropy",metrics=['accuracy'])
```

```python
es=EarlyStopping(monitor='val_accuracy',mode='max',patience=10,restore_best_weights=True)


```

```python
history=model.fit(X,to_categorical(Y),callbacks=[es],epochs=15,batch_size=15,validation_split=0.2,shuffle=True,verbose=1)
```

```python
history_dict=history.history
```

```python
loss_values=history_dict['loss']
val_loss_values=history_dict['val_loss']
```

```python
epochs=range(1,len(loss_values)+1)
```

```python
plt.plot(epochs,loss_values,'bo',label='Training loss')
plt.plot(epochs,val_loss_values,'orange',label='validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.show()
```

```python
acc=history_dict['accuracy']
val_acc=history_dict['val_accuracy']
epochs=range(1,len(acc)+1)


plt.plot(epochs,acc,'bo',label='Taining accuracy')
plt.plot(epochs,val_acc,'orange',label='validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.show()
```
